```python
import os
import csv
import random
from pathlib import Path
from tqdm import tqdm
import shutil
import cv2
import numpy as np
from PIL import Image
import albumentations as A

# ---------------- CONFIG ----------------
SRC_ROOT = Path("/content/khcrds")   # original dataset folder (each class = subfolder)
OUT_ROOT = Path("/content/Kannada_500_balanced")   # output balanced dataset
TARGET_PER_CLASS = 500
RANDOM_SEED = 42
IMG_EXTS = {".jpg", ".jpeg", ".png", ".bmp", ".webp"}
# ----------------------------------------
OUT_ROOT.mkdir(parents=True, exist_ok=True)
random.seed(RANDOM_SEED)
np.random.seed(RANDOM_SEED)
```

```python
augment = A.Compose([
    A.Resize(64, 64),  # change size if needed
    A.Rotate(limit=10, p=0.5),
    A.RandomBrightnessContrast(brightness_limit=0.1, contrast_limit=0.1, p=0.4),
    A.HorizontalFlip(p=0.2),
    A.GaussianBlur(blur_limit=3, p=0.2),
])
```

```python
def read_image(path):
    img = cv2.imread(str(path))
    if img is None:
        raise IOError(f"Failed to read image: {path}")
    return img

def save_image(img, path):
    path.parent.mkdir(parents=True, exist_ok=True)
    cv2.imwrite(str(path), img)
```

```python
class_to_images = {}

for cls_dir in sorted(SRC_ROOT.iterdir()):
    if cls_dir.is_dir():
        imgs = [p for p in cls_dir.iterdir() if p.suffix.lower() in IMG_EXTS]
        if imgs:
            class_to_images[cls_dir.name] = imgs

print(f"‚úÖ Found {len(class_to_images)} classes in dataset.")
for k, v in list(class_to_images.items())[:5]:
    print(f"  {k}: {len(v)} images")
```

```python
generated_rows = []

for cls, img_paths in tqdm(sorted(class_to_images.items()), desc="Processing classes"):
    out_cls_dir = OUT_ROOT / cls
    out_cls_dir.mkdir(parents=True, exist_ok=True)

    # 1Ô∏è‚É£ Copy all originals first
    for img_p in img_paths:
        dst = out_cls_dir / img_p.name
        shutil.copyfile(img_p, dst)
        generated_rows.append([str(dst.relative_to(OUT_ROOT)), cls])

    # 2Ô∏è‚É£ If fewer than target, augment until we reach 500
    current_count = len(img_paths)
    while current_count < TARGET_PER_CLASS:
        src = random.choice(img_paths)
        try:
            img = read_image(src)
        except:
            continue

        aug_img = augment(image=img)["image"]
        out_name = f"{src.stem}_aug_{current_count:04d}.jpg"
        out_path = out_cls_dir / out_name
        save_image(aug_img, out_path)
        generated_rows.append([str(out_path.relative_to(OUT_ROOT)), cls])
        current_count += 1

    print(f"‚úÖ {cls}: total {current_count} images")
```

```python
OUT_CSV = OUT_ROOT / "kannada_metadata_500.csv"
with OUT_CSV.open("w", newline="", encoding="utf-8") as f:
    writer = csv.writer(f)
    writer.writerow(["image_path", "class_label"])
    writer.writerows(generated_rows)

print(f"üìÅ New metadata CSV saved at: {OUT_CSV}")
```

```python
# Install if needed (Colab)
!pip install -q albumentations==1.3.0
!pip install -q timm

# Imports
import os
from pathlib import Path
import random
import csv
from glob import glob
from collections import Counter

import numpy as np
from PIL import Image
import cv2
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader

from sklearn.model_selection import train_test_split
import albumentations as A
from albumentations.pytorch import ToTensorV2
from tqdm import tqdm
```

```python
class CFG:
    BALANCED_ROOT = Path("/content/Kannada_500_balanced")  # folder with class subfolders
    CSV_PATH = Path("/content/kannada_metadata_500.csv")   # optional - we'll build from folders if absent
    IMAGE_SIZE = 48           # input size for CNN (keep >=32)
    BATCH_SIZE = 32
    NUM_WORKERS = 2           # reduce if Colab warns
    NUM_EPOCHS = 10
    LR = 1e-3
    DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    RANDOM_SEED = 42
    TEST_SIZE = 0.1
    VAL_SIZE = 0.1            # fraction of remaining after test (i.e. 0.1 -> 10% val)
    MODEL_SAVE_PATH = "kannada_resnet_bilstm_attn.pth"

cfg = CFG()
random.seed(cfg.RANDOM_SEED)
np.random.seed(cfg.RANDOM_SEED)
torch.manual_seed(cfg.RANDOM_SEED)
```

```python
# Build rows [image_path_relative_to_root, class_label_name]
rows = []
img_exts = {".jpg", ".jpeg", ".png", ".bmp", ".webp"}

for cls_dir in sorted(cfg.BALANCED_ROOT.iterdir()):
    if not cls_dir.is_dir():
        continue
    for img in sorted(cls_dir.iterdir()):
        if img.suffix.lower() in img_exts:
            rows.append([str(img.relative_to(cfg.BALANCED_ROOT)), cls_dir.name])

print(f"Found {len(rows)} images across {len(set(r[1] for r in rows))} classes")

# Save CSV
with open(cfg.CSV_PATH, "w", newline="", encoding="utf-8") as f:
    writer = csv.writer(f)
    writer.writerow(["image_path", "class_label"])
    writer.writerows(rows)
print("Saved metadata to:", cfg.CSV_PATH)
```

```python
# Read CSV
data = []
with open(cfg.CSV_PATH, newline="", encoding="utf-8") as f:
    reader = csv.DictReader(f)
    for r in reader:
        data.append((r["image_path"], r["class_label"]))

paths, labels = zip(*data)
unique_classes = sorted(list(set(labels)))
num_classes = len(unique_classes)
print("Classes:", num_classes)

# class -> int
cls2idx = {c: i for i, c in enumerate(unique_classes)}
idx2cls = {i: c for c, i in cls2idx.items()}

# convert to arrays for sklearn
paths = np.array(paths)
labels = np.array([cls2idx[l] for l in labels])

# first split out test
paths_trainval, paths_test, labels_trainval, labels_test = train_test_split(
    paths, labels, test_size=cfg.TEST_SIZE, stratify=labels, random_state=cfg.RANDOM_SEED
)

# then split train/val
val_frac_of_trainval = cfg.VAL_SIZE / (1.0 - cfg.TEST_SIZE)
paths_train, paths_val, labels_train, labels_val = train_test_split(
    paths_trainval, labels_trainval, test_size=val_frac_of_trainval,
    stratify=labels_trainval, random_state=cfg.RANDOM_SEED
)

print("Train:", len(paths_train), "Val:", len(paths_val), "Test:", len(paths_test))
```

```python
class KannDataset(Dataset):
    def __init__(self, root:Path, rel_paths, labels, transform=None):
        self.root = Path(root)
        self.rel_paths = list(rel_paths)
        self.labels = list(labels)
        self.transform = transform

    def __len__(self):
        return len(self.rel_paths)

    def __getitem__(self, idx):
        rel = self.rel_paths[idx]
        img_path = self.root / rel
        # read with cv2 (BGR) then convert to RGB
        img = cv2.imread(str(img_path))
        if img is None:
            raise IOError(f"Cannot read image: {img_path}")
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        if self.transform:
            img = self.transform(image=img)["image"]
        label = self.labels[idx]
        return img, label

# Transforms
train_transform = A.Compose([
    A.Resize(cfg.IMAGE_SIZE, cfg.IMAGE_SIZE),
    A.Rotate(limit=10, p=0.5),
    A.ShiftScaleRotate(shift_limit=0.03, scale_limit=0.05, rotate_limit=0, p=0.3),
    A.RandomBrightnessContrast(0.1, 0.1, p=0.4),
    A.HorizontalFlip(p=0.2),
    A.Normalize(mean=(0.485,0.456,0.406), std=(0.229,0.224,0.225)),
    ToTensorV2()
])

val_transform = A.Compose([
    A.Resize(cfg.IMAGE_SIZE, cfg.IMAGE_SIZE),
    A.Normalize(mean=(0.485,0.456,0.406), std=(0.229,0.224,0.225)),
    ToTensorV2()
])
```

```python
train_ds = KannDataset(cfg.BALANCED_ROOT, paths_train, labels_train, transform=train_transform)
val_ds = KannDataset(cfg.BALANCED_ROOT, paths_val, labels_val, transform=val_transform)
test_ds = KannDataset(cfg.BALANCED_ROOT, paths_test, labels_test, transform=val_transform)

train_loader = DataLoader(train_ds, batch_size=cfg.BATCH_SIZE, shuffle=True, num_workers=cfg.NUM_WORKERS, pin_memory=True)
val_loader = DataLoader(val_ds, batch_size=cfg.BATCH_SIZE, shuffle=False, num_workers=cfg.NUM_WORKERS, pin_memory=True)
test_loader = DataLoader(test_ds, batch_size=cfg.BATCH_SIZE, shuffle=False, num_workers=cfg.NUM_WORKERS, pin_memory=True)
```

```python
import torchvision.models as models

class ResNetBiLSTMAttn(nn.Module):
    def __init__(self, num_classes, cnn_out_dim=512, lstm_hidden=256, lstm_layers=1, pretrained=False):
        super().__init__()
        # ResNet backbone
        self.backbone = models.resnet18(weights=models.ResNet18_Weights.IMAGENET1K_V1 if pretrained else None)
        # remove fc and avgpool -> we'll produce feature vector of size cnn_out_dim
        self.backbone.fc = nn.Identity()
        # Note: resnet18 outputs 512-d features
        self.cnn_out_dim = cnn_out_dim

        # LSTM (we'll treat CNN feat as sequence length 1 -> BiLSTM will return same dim*2)
        self.lstm = nn.LSTM(self.cnn_out_dim, lstm_hidden, num_layers=lstm_layers,
                            bidirectional=True, batch_first=True)
        self.attn = nn.Linear(lstm_hidden*2, 1)   # attention score
        self.classifier = nn.Linear(lstm_hidden*2, num_classes)

    def forward(self, x):
        # x: B x 3 x H x W
        feat = self.backbone(x)         # B x 512
        feat = feat.unsqueeze(1)        # B x seq_len(=1) x 512
        lstm_out, _ = self.lstm(feat)   # B x 1 x (2*h)
        # Attention
        scores = self.attn(lstm_out)    # B x 1 x 1
        weights = torch.softmax(scores, dim=1)  # B x 1 x 1 (trivial but consistent)
        context = (weights * lstm_out).sum(dim=1)  # B x (2*h)
        out = self.classifier(context)
        return out

model = ResNetBiLSTMAttn(num_classes=num_classes, pretrained=True).to(cfg.DEVICE)
print("Model params:", sum(p.numel() for p in model.parameters() if p.requires_grad))
```

```python
criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters(), lr=cfg.LR)
scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(optimizer, mode='min', factor=0.5, patience=3)
```

```python
def train_one_epoch(model, loader, criterion, optimizer, device):
    model.train()
    running_loss = 0.0
    correct = 0
    total = 0
    for imgs, labels in tqdm(loader, desc="Train"):
        imgs = imgs.to(device)
        labels = labels.to(device)
        optimizer.zero_grad()
        outputs = model(imgs)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()
        running_loss += loss.item() * imgs.size(0)
        _, preds = outputs.max(1)
        correct += preds.eq(labels).sum().item()
        total += labels.size(0)
    return running_loss/total, 100.0 * correct/total

def validate(model, loader, criterion, device):
    model.eval()
    running_loss = 0.0
    correct = 0
    total = 0
    with torch.no_grad():
        for imgs, labels in tqdm(loader, desc="Val"):
            imgs = imgs.to(device)
            labels = labels.to(device)
            outputs = model(imgs)
            loss = criterion(outputs, labels)
            running_loss += loss.item() * imgs.size(0)
            _, preds = outputs.max(1)
            correct += preds.eq(labels).sum().item()
            total += labels.size(0)
    return running_loss/total, 100.0 * correct/total
```

```python
best_val_acc = 0.0
for epoch in range(cfg.NUM_EPOCHS):
    print(f"\n=== Epoch {epoch+1}/{cfg.NUM_EPOCHS} ===")
    train_loss, train_acc = train_one_epoch(model, train_loader, criterion, optimizer, cfg.DEVICE)
    val_loss, val_acc = validate(model, val_loader, criterion, cfg.DEVICE)
    print(f"Train loss: {train_loss:.4f} | Acc: {train_acc:.2f}%")
    print(f"Val   loss: {val_loss:.4f} | Acc: {val_acc:.2f}%")
    scheduler.step(val_loss)
    if val_acc > best_val_acc:
        best_val_acc = val_acc
        torch.save({
            "model_state_dict": model.state_dict(),
            "optimizer_state_dict": optimizer.state_dict(),
            "epoch": epoch,
            "cls2idx": cls2idx
        }, cfg.MODEL_SAVE_PATH)
        print(f"Saved best model (val acc {best_val_acc:.2f}%) -> {cfg.MODEL_SAVE_PATH}")
```
